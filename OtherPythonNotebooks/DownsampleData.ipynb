```python
import os
from pathlib import Path
from tifffile import imread, imwrite
import cv2
import glob
```

```python
def DownsampleData(image, DownsampleFactor):

                if DownsampleFactor != 1:
                    print('Downsampling Image in XY by', DownsampleFactor)
                    # percent of original size
                    scale_percent=int(100/DownsampleFactor)
                    width=int(image.shape[1] * scale_percent / 100)
                    height=int(image.shape[0] * scale_percent / 100)
                    dim=(width, height)
                    smallimage=cv2.resize(
                              image.astype('float32'), dim)

                    return smallimage
               
```

```python
imagedir = '/Users/rando/Downloads/NLM-MontgomeryCXRSet/MontgomerySet/ManualMask/BinaryMask/'
savedir = '/Users/rando/Downloads/NLM-MontgomeryCXRSet/MontgomerySet/ManualMask/dsBinaryMask/'
```

```python
Raw_path = os.path.join(imagedir, '*tif')
X = glob.glob(Raw_path)
Path(savedir).mkdir(exist_ok=True)
for fname in X:
    Name = os.path.basename(os.path.splitext(fname)[0])
    image = imread(fname)
    dsimage = DownsampleData(image, 4)
    imwrite(savedir + '/' + Name + '.tif', dsimage.astype('float32'))
```
