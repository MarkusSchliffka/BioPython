```python
sourcedirKymo = '/Users/aimachine/DarineKymo/'
savedir = sourcedirKymo + 'Results/'
Time_unit = 1
Xcalibration = 1
```

```python
Raw_path = os.path.join(sourcedirKymo, '*tif')
X = glob.glob(Raw_path)
Path(savedir).mkdir(exist_ok = True)
AverageVelocity = []
AverageFrequ = []
Frequency = []
ImageName = []
for fname in X:
    image = imread(fname)
    image = np.swapaxes(image,0,1)
    
    #X axis should be time
    print(image.shape)
    Name = os.path.basename(os.path.splitext(fname)[0])
    #To find the frequency of oscillations
    FFTImage = FFTStrip(image)
    doubleplot(image, FFTImage, "Original Image",  "FFT")
    imwrite((savedir + "TwoDFFT" + Name + '.tif') , FFTImage.astype('float32'))
    onedimg, frequ = doFilterFFT(image, Time_unit, False)
    plt.plot(frequ, onedimg, label = 'Frequency')
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.show()
    
    frqY = show_peak(onedimg, frequ, 0) 
    AverageFrequ.append(frqY)
    #Get simple velocity
    time = range(0, image.shape[1])
    time = np.asarray(time)
  
    simplevelocity = VelocityStrip(image, Xcalibration, Time_unit)
    AverageVelocity.append(np.mean(simplevelocity))
    ImageName.append(Name)
    plt.plot(time, simplevelocity, label = 'Velocity')
    plt.xlabel('Time')
    plt.ylabel('Velocity')
    plt.show()

df = pd.DataFrame(list(zip(AverageVelocity,AverageFrequ)), index = ImageName, columns =['Velocity', 'Frequency'])
df.to_csv(savedir + Name + "SimpleVelocity"  '.csv', index = False)  
df    

```
