```python
import os
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import curve_fit
from lmfit import Model
from numpy import exp, loadtxt, pi, sqrt
```

```python
CsvDir = '/home/kapoorlab/Downloads/Darine/Nucleus_tracking/'
Savedir = '/home/kapoorlab/Downloads/Darine/Nucleus_tracking/Results/'
```

```python

def gaussian(x, amp, mu, std):
    return amp * exp(-(x-mu)**2 / std)
```

```python
Raw_path = os.path.join(CsvDir, '*csv')
X = glob.glob(Raw_path)
for CsvFile in X:
  print(CsvFile)  
  dataset = pd.read_csv(CsvFile)
  df = pd.DataFrame(dataset)
    
  displacement = dataset["Distance"][1:]
  displacement = np.asarray(displacement)
  time = dataset["Slice"][1:]  
  time = np.asarray(time)
    
  plt.plot(time, displacement)   

  plt.xlabel("Time")
  plt.ylabel("Displacement")
  plt.show()   
  

  mean, std = norm.fit(displacement) 
  
  counts, bins = np.histogram(displacement, bins = 20)
  bins = bins[:-1]  
  gmodel = Model(gaussian)
  Gauss = gmodel.fit(counts, x=bins, amp=np.max(counts), mu=mean, std=std)
  plt.hist(bins, bins, weights=counts)
  plt.plot(bins, Gauss.best_fit)
  plt.xlabel("Delta")
  plt.ylabel("Counts")
  plt.show()  
  print(Gauss.fit_report())
  
```
