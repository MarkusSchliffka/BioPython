```python
import os
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
CsvDir = '/home/kapoorlab/Downloads/Darine/Nucleus_tracking/'
Savedir = '/home/kapoorlab/Downloads/Darine/Nucleus_tracking/Results/'
```

```python
Raw_path = os.path.join(CsvDir, '*csv')
X = glob.glob(Raw_path)
for CsvFile in X:
  dataset = pd.read_csv(CsvFile)
  df = pd.DataFrame(dataset)
  print(dataset.shape)
    
  displacement = dataset["Distance"]
  displacement = np.asarray(displacement)
  counts, bins = np.histogram(displacement)
  plt.hist(bins[:-1], bins, weights=counts)
  plt.show()
  print(CsvFile)
```
